[ [ corr(x, x), corr(x, y) ],
 [ corr(y, x), corr(y, y) ]

```python
np.corrcoef(x, y)
```

create subplots() function from matplot.pyplot to create a figure and the axes onto which we plot our data

```python
from matplotlib.pyplot import subplots
```

```python
rng = np.random.default_rng()
```

```python
fig, ax = subplots(figsize = ( 8, 8 ))
x = rng.standard_normal(100)
y = rng.standard_normal(100)
ax.plot(x, y);
```

```python
fig, ax = subplots(figsize = (8, 8))
ax.scatter(x, y, marker = "*")
```

```python
fig, ax = subplots(figsize = (8, 8))
ax.scatter(x, y, marker = "o")
ax.set_xlabel("X AXIS")
ax.set_ylabel("Y AXIS")
ax.set_title("SCATTER PLOT")
```

CHANGE THE SIZE FROM 8,8 TO 12,3

```python
fig.set_size_inches(12,3)
fig
```

```python
fig, axes = subplots(
    nrows = 3,
    ncols = 3,
    figsize = (12, 3)
)
axes[0,1].plot(x, y, 'o')
axes[1,2].scatter(x, y, marker = "+")
```

USE LINSPACE(A, B, N) => TO RETURN N NUMBERS FROM A TO B

```python
fig, ax = subplots(figsize = (8, 8))
n = np.linspace(-np.pi, np.pi, 50)
n
y = x # form a straight line
# outer function returns the product of two transformed arrays y and x
# res[i, j] = a[i] * b[j]
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))
ax.contour(x, y, f, levels = 45)
```
