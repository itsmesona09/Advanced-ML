**1. For each of parts (a) through (d), indicate whether we would generally
expect the performance of a flexible statistical learning method to be
better or worse than an inflexible method. Justify your answer.**

**(a) The sample size n is extremely large, and the number of predictors p is small.**

- A flexible method performs better when the sample size is large and the number of predictors is small.
- Overfitting is less of a concern in this case, as there is enough data to learn the true patterns and not too many features to get distracted by.
- A flexible model can achieve lower bias without significantly increasing variance.

**(b) The number of predictors p is extremely large, and the number of observations n is small.**

- Inflexible method performs better when there is a large number of predictors and the number of observations is small.
- A flexible model is prone to overfitting when the number of predictors is large and the number of observations is small.
- It can fit the training data perfectly, including the noise but this leads to high variance and poor generalization to unseen data, especially when using complex or non-linear models.

**(c) The relationship between the predictors and response is highly
non linear.**

- A flexible method performs better when the relationship between the predictors and the response is non-linear.
- It can adapt to the complex structure of the data and provide better predictive accuracy compared to inflexible methods, which may have high bias in such scenarios.

**(d) The variance of the error terms, i.e. Ïƒ2 = Var(e), is extremely
high.**

- Inflexible method performs better when the variance is extremely high. When there's a lot of noise in the output variable, there is a high variance in the error terms.
- Flexible methods are more likely to capture this noise, leading to high variance in their predictions.
- Inflexible methods can give more stable predictions and can perform in high noise settings.

**3. We now revisit the bias-variance decomposition.**

**(a) Provide a sketch of typical (squared) bias, variance, training er-
ror, test error, and Bayes (or irreducible) error curves, on a sin-
gle plot, as we go from less flexible statistical learning methods towards more flexible approaches. The x-axis should represent the amount of flexibility in the method, and the y-axis should represent the values for each curve. There should be five curves. Make sure to label each one.**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)

bias2 = (1 - x) ** 2
variance = x ** 2
training_error = 0.2 - 0.15 * x
test_error = bias2 + variance + 0.1
irreducible_error = np.ones_like(x) * 0.1
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, bias2, label='BiasÂ²')
plt.plot(x, variance, label='Variance')
plt.plot(x, training_error, label='Training Error')
plt.plot(x, test_error, label='Test Error')
plt.plot(x, irreducible_error, label='Bayes Error (Irreducible)', linestyle='dashed')

plt.xlabel('Model Flexibility')
plt.ylabel('Error / Variance')
plt.title('Bias-Variance Tradeoff')
plt.legend()
plt.grid(True)
plt.show()
```

**(b) Explain why each of the five curves has the shape displayed in part (a).**

**1. BIAS^2**

It decreases as the flexibility increases.

Inflexible models like Linear Regression make strong assumptions (High Bias).

Flexible models adapt to better data (Low Bias).

**2. VARIANCE**

It increases as the variance increases.

Inflexible models are stable (Low Variance).

Flexible models change a lot with small changes in the data (High Variance).

**3. TRAINING ERROR**

It steadily decreases as the model flexibility increases.

More flexibility models can fit the training data more closely.

Therefore, the training error goes down with flexibility.

**4. TEST ERROR**

The Test error is U shaped.

It initially reduces as bias reduces, then increases as variance increases.

Minimum point is the best generalization.

**5. BAYES ERROR**

It is a flat line.

This error comes from noise in the data, which no model can reduce.

It is called irreducible error.

```python
auto = pd.read_csv('Auto.csv', na_values='?')
auto = auto.dropna()

auto_subset = auto.drop(auto.index[9:85])
quantitative_cols = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year']
summary_stats = {}
for col in quantitative_cols:
    col_data = auto_subset[col]
    col_range = (col_data.min(), col_data.max())
    col_mean = col_data.mean()
    col_std = col_data.std()
    summary_stats[col] = {
        'range': col_range,
        'mean': col_mean,
        'std_dev': col_std
    }

import pprint
pprint.pprint(summary_stats)
```
